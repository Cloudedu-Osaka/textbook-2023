### Êï∞Â≠ó„ÅÆÂ§âÊèõËæûÊõ∏„Çí‰Ωú„ÇäÁøªË®≥„Åô„ÇãÈñ¢Êï∞„ÇíÂÆöÁæ©„Åô„Çã

```python
babylonian_numerals = {}
for i in range(1, 10):
    # print(i, chr(ord('\U00012415')-1+i))
    babylonian_numerals[i] = chr(ord('\U00012415')-1+i)

babylonian_numerals[10] = u'\U0001230B'
babylonian_numerals[20] = u'\U00012399'
babylonian_numerals[30] = u'\U0001230D'
babylonian_numerals[40] = u'\U0001240F'
babylonian_numerals[50] = u'\U00012410'

babylonian_numerals
```

```python
def decimal_to_babylonian(num):
    result = ''
    for value in sorted(babylonian_numerals.keys(), reverse=True):
        while num >= value:
            result += babylonian_numerals[value]
            num -= value
    return result
```

### 10ÈÄ≤Êï∞„Çí60ÈÄ≤Êï∞„Å´Â§âÊèõ„Åô„ÇãÈñ¢Êï∞„ÇíÂÆöÁæ©„Åô„Çã

```python
%%ai_ask
python„Åß10ÈÄ≤Êï∞„ÅÆÂÆüÊï∞„Çí60ÈÄ≤Êï∞„Å´Â§âÊèõ„Åô„Çã„ÄÇ„Åü„Å†„Åó„ÄÅ0„Åã„Çâ59„Åæ„Åß„ÅÆÊï∞Â≠ó„ÅÆ„É™„Çπ„Éà„ÇíÂá∫Âäõ„Åô„Çã„ÄÇ
```

```python
def decimal_to_sexagesimal(decimal):
    integer_part = int(decimal)
    fractional_part = decimal - integer_part
    sexagesimal_integer = []
    while integer_part > 0:
        remainder = integer_part % 60
        sexagesimal_integer.append(remainder)
        integer_part //= 60
    sexagesimal_integer.reverse()
    sexagesimal_fractional = []
    while fractional_part > 0 and len(sexagesimal_fractional) < 10:
        product = fractional_part * 60
        integer_part = int(product)
        sexagesimal_fractional.append(integer_part)
        fractional_part = product - integer_part
    return sexagesimal_integer, sexagesimal_fractional
```
